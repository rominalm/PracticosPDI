```python
#### propiedades ####
def propiedades(image):
  plt.imshow(image)
  plt.figure()

  print ("La imagen tiene", image.shape[0], "columnas,", image.shape[1], "filas y",  image.shape[2], "canales")
  print ("El tipo de dato de la imagen es", image.dtype)

  rows = image.shape[0]
  columns = image.shape[1]

  return(rows, columns)
```

```python
#### visualizar cada canal ####
def chanels(image):
  copy_im = image.copy()
  for i in range(0, 3):
    if i == 0:
      save_name = 'image_red.jpg'
    if i== 1:
      save_name = 'image_green.jpg'
    if i== 2:
      save_name = 'image_blue.jpg'
    plt.imshow(copy_im[:,:,i])
    plt.title(save_name)
    plt.figure()
    imageio.imsave(save_name, copy_im[:,:,i])
```

```python
#### cambiar color de columnas de pixeles ####
def changepixel1(image, rows, columns):
  copy_im = image.copy()
  for j in range(0, columns, 20):
    for i in range(0, rows):
      copy_im[i, j] = (255, 0, 0)
  plt.imshow(copy_im)
  plt.title("image color lines")
  plt.figure()
  imageio.imsave('image_lines.jpg', copy_im)
```

```python
#### cambiar color de pixeles ####
def changepixel2(image, rows, columns):
  copy_im = image.copy()
  for k in range(0, rows):
    for l in range(0, columns):
      pix = copy_im[k, l]
      if pix[0] < 100 and pix[2] == 0:
        copy_im[k, l] = (255, 0, 255)
  plt.imshow(copy_im)
  plt.title("image violet")
  plt.figure()
  imageio.imsave('image_violet.jpg', copy_im)
```

```python
#### abrir imagen ####
image = imageio.imread("/content/gdrive/My Drive/PDI-TP/girasol.jpg")

rows, columns = propiedades(image)
chanels(image)
changepixel1(image, rows, columns)
changepixel2(image, rows, columns)
```
